```python
# 排序之后，进行累计求和
data = data.sort_index()
```

```python
stock_rise = data['p_change']
# plot方法集成了前面直方图、条形图、饼图、折线图
# stock_rise.cumsum()
```

```python
import matplotlib.pyplot as plt
# plot显示图形
stock_rise.cumsum().plot()
# 需要调用show，才能显示出结果
plt.show()
```
